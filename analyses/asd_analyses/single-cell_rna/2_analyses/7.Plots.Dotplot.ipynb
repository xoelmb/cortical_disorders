```r
filter.de.results <- function(df,
                              pval.col, 
                              logfc.col,
                              use.abs.logfc=FALSE,
                              
                              max.pval=0.05, 
                              min.logfc=log2(1),
                              
                              top.by='pval', 
                              top.n=50, 
                              ties.method='first',
                              add.pars = T,
                              split.by=NULL){
    
    # split analysis
    if (!is.null(split.by)){
        
        split(df, df[,split.by]) %>%
        lapply(FUN = function(d){
            filter.de.results(df=d,
                              pval.col=pval.col, 
                              logfc.col=logfc.col,
                              use.abs.logfc=use.abs.logfc,
                              max.pval=max.pval,
                              min.logfc=min.logfc,
                              top.by=top.by,
                              top.n=top.n,
                              ties.method=ties.method,
                              add.pars=add.pars)
        }) %>% do.call(what='rbind') %>%
        as.data.frame %>% return
        
    } else {
        
        # 1. set columns 
        df['pval.use'] <- df[,pval.col]
        df['logfc.use'] <- df[,logfc.col]
        # 2. Only positive markers?
        if (use.abs.logfc) {df['logfc.used'] <- abs(df['logfc.used'])}
        # 3. add filters 
        df <- df %>% mutate(
            pval.isnumeric = is.numeric(pval.use)&!is.na(pval.use),
            logfc.isnumeric = is.numeric(logfc.use)&!is.na(logfc.use),
            # 4. thresholds
            pval.underormax = pval.use <= max.pval,
            logfc.overormin = logfc.use >= min.logfc,
            # 5. obtain valid values to rank
            rank.consider = pval.isnumeric&logfc.isnumeric&pval.underormax&logfc.overormin,
            pval.only_valid = ifelse(rank.consider, pval.use, NA),
            logfc.only_valid = ifelse(rank.consider, logfc.use, NA),
            # 6. rank values
            pval.rank = rank(pval.only_valid, na.last = T, ties.method = ties.method),
            logfc.rank = rank(-logfc.only_valid, na.last = T, ties.method = ties.method)
        )
        # get top
        df$pval.top <- df$pval.rank <= top.n
        df$logfc.top <- df$logfc.rank <= top.n
        # combine filters
        df$all.ok <- df$rank.consider & df[,paste0(top.by, '.top')]
        
        if (add.pars){
            df %>% mutate(
                pval.col=pval.col, 
                logfc.col=logfc.col,
                use.abs.logfc=use.abs.logfc,
                max.pval=max.pval,
                min.logfc=min.logfc,
                top.by=top.by,
                top.n=top.n,
                ties.method=ties.method)
        }
        
        return(df)
        
    }
}
```

```r
scale_zero_to_max <- function(x){
    
    if (any(x<0)){stop('Does not work with negative numbers')}
    
    max.x <- max(x, na.rm=T)
    
    # If everything is zero, return original
    if (max.x == 0){
        return(x)
    } else {
        # otherwise
        return(x/max.x)
    }
}
scale_min_to_max <- function(x){
        
    max.x <- max(x, na.rm=T)
    if (max.x == 0){
        return(x)
    } else {
        min.x <- min(x)
        # otherwise
        return((x-min.x)/(max.x-min.x))
    }
}

labels_zero_to_max <- function(x){
    # message(x)
    y <- scales::label_percent()(x)
    # message(y)
    y[length(y)] <- 'Max'
    return(y)
}

labels_min_to_max <- function(x){
    # messsage(x)
    y <- scales::label_percent()(x)
    # messsage(y)
    y[1] <- 'Min'
    y[length(y)] <- 'Max'
    return(y)
}
```

```r
dotplot <- function(seu, gene.lists, 
                    repeated.do = 'label', force_hide_repeated=T,
                    exp.slot = 'scale.data',
                    
                    min.col='white', max.col='orangered3',
                    min.point.size=.1, max.point.size=3, 
                    
                    pct.min=3, n.samples.over.pct=1,
                    
                    clip.upper.pct=100, n.breaks.pct=6,
                    
                    pct_scale_by=NULL, pct_scaler = 'zero_to_max',
                    exp_scale_by=NULL, exp_scaler = 'min_to_max',
                    
                    cluster.genes=T, max.clusters=6, cluster_on='pct', 
                    
                    title=NULL, subtitle=NULL, y.lab='genes', 
                    ...){

    require(ggh4x)
    facet.vars <- vars(feature.groups)
    
    exp.is_scaled <- F
    pct.is_scaled <- F
    pct_labeller <- function(...){return(as.character(...))}
    exp_labeller <- function(...){return(as.character(...))}
    
    
    ## PROCESS WITH REPEATED GENES
    repeated.genes <- list()
    nr.genes <- list()
    for (l in names(gene.lists)){
        clist <- gene.lists[names(gene.lists)!=l]
        repeated <- gene.lists[[l]]%in%unique(unlist(clist))
        print(glue::glue('{sum(repeated)} ({round(sum(repeated)/length(gene.lists[[l]]), 4)*100}%) repeated genes in {l} list'))
        nr.genes[[l]] <- gene.lists[[l]][!repeated]
        repeated.genes[[l]] <- gene.lists[[l]][repeated]
    }
    # remove empty lists
    nr.genes <- nr.genes[sapply(nr.genes, length)!=0]
    
    # REMOVE OR ADD
    if ((repeated.do == 'label') & (sum(sapply(repeated.genes, length))>0)){
        nr.genes$repeated <- unique(unlist(repeated.genes))
        facet.vars <- vars(feature.groups, repeated)
    } else if (!repeated.do%in%c('remove', 'label')){
        message(glue::glue('"{repeated.do}" is not a valid `repeated.do` option ["remove", "label"]. Removing...'))
    }
    
    
    # ANY VALID GENES ?
    nplot.genes <- sum(sapply(nr.genes, length)) 
    if (nplot.genes==0){return(NULL)}
    
    # FETCH DATA FROM DOTPLOT
    p <- DotPlot(seu, 
                 group.by='anno_cluster_fct', split.by='cell_origin', 
                 features=nr.genes, cols = 'Reds', ...)
    
    pdata <- p$data$id%>% as.character%>% strsplit(split='_D8_', fixed=T) %>%
        as.data.frame() %>% t()%>% as.data.frame() %>%
        mutate(celltype=factor(V1, cell.types.nohighmito), 
               Donor=factor(paste0('D8_', V2), samples)) %>%
        cbind(p$data, .) %>% as.data.frame()
    
    
    # REMOVE LOW EXP BASED ON PCT
    if ( is.numeric(pct.min)&is.numeric(n.samples.over.pct)&(n.samples.over.pct>0)&(pct.min>0) ){
        message(glue::glue('Requiring {pct.min}% expression in at least {n.samples.over.pct} group(s)'))
        
        (pdata%>% 
         mutate(pct.over=pct.exp>pct.min)%>% group_by(features.plot) %>% summarise(n.pct.over = sum(pct.over)) %>%  subset(n.pct.over>=n.samples.over.pct)
        )$features.plot -> exp.genes
        
        
        print(glue::glue('Number of genes {length(unique(pdata$features.plot))} -> {length(exp.genes)}'))
        pdata <- subset(pdata, features.plot %in% exp.genes)
        
    } 
    # ADD CORRESPONDING REPEATED GENES
    if ((repeated.do == 'label')&(sum(sapply(repeated.genes, length))>0)){
        p.repeated <- subset(pdata, feature.groups == 'repeated')
        pdata <- subset(pdata, feature.groups != 'repeated') 

        lapply(names(repeated.genes), function(x){
            subset(p.repeated, features.plot %in% repeated.genes[[x]])%>%mutate(feature.groups=factor(x, names(gene.lists)), repeated='Not exclusive')
        }) %>% do.call(what='rbind') %>% rbind(pdata%>%mutate(repeated='Exclusive'))%>%
        mutate(repeated=factor(repeated, c('Exclusive', 'Not exclusive'))) -> pdata

    } else {
        pdata$repeated <- factor('Exclusive')
    }
    
    
    
    
    # N GENES UNIQUE
    nunique.genes <- length(unique(pdata$features.plot))
    
    # PROCESS EXPRESSION
    # Define expression slot
    if(exp.slot=='data'){
        pdata$exp <- pdata$avg.exp
    } else if (exp.slot == 'scale.data'){
        pdata$exp <- pdata$avg.exp.scaled
        exp.is_scaled <- T
    }
    # CLIP PCT
    pdata[pdata$pct.exp>clip.upper.pct,'pct.exp'] <- clip.upper.pct 

    # scales
    if (!is.null(pct_scale_by)){
        pct_labeller <- if(pct_scaler=='zero_to_max'){labels_zero_to_max}else{labels_min_to_max}
        pct_scaler <- if(pct_scaler=='zero_to_max'){scale_zero_to_max}else{scale_min_to_max}
        
        if (clip.upper.pct!=100){warning('Clipping and scaling pct is not advised')}
        pdata <- pdata%>%group_by_at(pct_scale_by) %>% mutate(pct.exp=pct_scaler(pct.exp)) 
        clip.upper.pct <- 1
    }
    if (!is.null(exp_scale_by)){
        exp.is_scaled <- T
        exp_labeller <- if(exp_scaler=='zero_to_max'){labels_zero_to_max}else{labels_min_to_max}
        exp_scaler <- if(exp_scaler=='zero_to_max'){scale_zero_to_max}else{scale_min_to_max}
        pdata <- pdata%>%group_by_at(pct_scale_by) %>% mutate(exp=exp_scaler(exp)) 
    }
    
    
    # CLUSTER GENES
    if (cluster.genes){
        if(nunique.genes>30){
            value.var <- if(cluster_on=='pct'){'pct.exp'}else{'exp'}
            exp.matrix <- reshape2::dcast(pdata, value.var = value.var,
                formula='feature.groups+repeated+features.plot~celltype+Donor')
            exp.matrix[is.na(exp.matrix)] <- min(exp.matrix[,-(1:3)], na.rm = T)

            require(amap)
            for (i in seq(from=min(max.clusters, nrow(exp.matrix)), to=1)){
                kms <- tryCatch({amap::Kmeans(x = exp.matrix[,-(1:3)], centers = i,iter.max = 100)},
                               error=function(...){NULL})
                if (!is.null(kms)){
                    pdata <- merge(pdata, cbind(exp.matrix[,1:3], cluster=kms$cluster), 
                                   by=c('feature.groups', 'repeated', 'features.plot'), all=T)
                    facet.vars <- if (repeated.do=='label'){
                        vars(cluster, feature.groups, repeated)}else{
                        vars(feature.groups, cluster)}    
                    break()
                }
            }
        }
    }
    

    
    # REMOVE REPEATED GROUP IF THERE IS NONE
    rep.fac <- grepl(pattern='repeated', 
                     gsub(pattern='~', replacement='', fixed=T, as.character(facet.vars)))
    if (any(rep.fac) & (force_hide_repeated | all(pdata$repeated=='Exclusive'))){
        facet.vars <- facet.vars[!rep.fac]
    }

    
    
    # PLOT
        # MAPPING, AXES, FACETS
    p <- ggplot(pdata, 
                aes(x=Donor, y=features.plot,
                    size = pct.exp, fill = exp)) +
        scale_x_discrete(position = 'top') + 
        facet_nested(rows=facet.vars, cols=vars(celltype),
                     strip = strip_nested(size = 'variable'), 
                     scales='free', space='free', switch='y') + 
        # DOTS
        geom_point(color='#222222FF', stroke=.05, shape=21) +
        # SCALES
        scale_fill_gradient(low = min.col, high=max.col, labels=exp_labeller)+
        scale_size(range=c(min.point.size, max.point.size),
                   labels=pct_labeller,
                   limits=c(0,clip.upper.pct), 
                   breaks=seq(from=0, to=clip.upper.pct, length.out = n.breaks.pct)) +
        # LEGEND
        guides(size=guide_legend(override.aes = list(fill=max.col), 
                                 title.position = 'top', nrow = 2),
               fill=guide_colorbar(title.position = 'top',
                                   direction = 'horizontal')) +
        # NAMES
        labs(y=y.lab, x='Celltype and donor',
             size=paste0(if(!is.null(pct_scale_by)){'Scaled '}else{''}, "Percent Expressed"), 
             fill=paste0(if(exp.is_scaled){'Scaled '}else{''}, "Average Expression")) + 
        # THEMING
        theme_linedraw() + 
        theme(panel.grid.major = element_line(color='#66666644', linewidth=0.2),
              strip.placement='outside',
              strip.text.x = element_text(color='black'),
              strip.text.y.left = element_text(color='black', angle=0),
              strip.background = element_rect(fill='#88888822'),
              axis.text.x.top=element_text(angle=90, hjust=0, vjust=0.5),
              legend.position='top')
    
    if(!is.null(title) | !is.null(subtitle)){
        p <- p + ggtitle(title, subtitle=subtitle)
    }
    
    return(p)

}
```

```r
dotplot.save <- function(
    seu, gene.list, name, save.dir='./', min.width = 0,
    with_clustering=T, cluster_on='pct', without_clustering=T,
    pct.min = 3, n.samples.over.pct = 1, repeated.do = 'label', genelevels=NULL, ...){
    
    if (any(sapply(gene.list, function(x){length(x)!=length(unique(x))}))){
        stop('check')
    }
    if (!any(with_clustering, without_clustering)){return()}
    
    
    
    dir.create(save.dir, recursive=T)
    message(name)
    
    ps <- list()
    
    # with clustering
    if (with_clustering){
        message('--> With clustering...')
        p <- dotplot(seu = seu, 
                     gene.list = gene.list, 
                     y.lab = name,
                     cluster_on=cluster_on,
                     cluster.genes = T,
                     pct.min = pct.min,
                     n.samples.over.pct = n.samples.over.pct, 
                     repeated.do = repeated.do,
                     ...)
        
        # override order of genes
        if (genelevels){
            glevels <- genelevels[[l]]
            glevels <- glevels[glevels%in%p$data$features.plot]
            p$data %>% mutate(features.plot = factor(as.character(features.plot), glevels)) -> p$data
        }
        
        if (!is.null(p)){
            n.groups <- p$data[,names(p$facet$params$rows), drop=F]%>%unique()
            n.groups <- if(!any(is.null(dim(n.groups)))){nrow(n.groups)}else{length(n.groups)}
            n.genes.plot <- nrow(unique(p$data[,c('features.plot', 'feature.groups')])) + n.groups*0.5

            n_cts <- length(unique(seu$anno_cluster_fct))
            
            if(n.genes.plot>=1){

                row.chars <- p$data[,names(p$facet$params$rows), drop=F]%>%
                    apply(1, function(x){sum(nchar(as.character(x)))}) %>% max
                if (row.chars==1){row.chars <- 10}

                pdim <- c(
                    # width = width.lm$coefficients['x']*row.chars+width.lm$coefficients[1], 
                    width = width.lm$coefficients['x']*row.chars+width.lm$coefficients[1]*n_cts/6+.5, 
                    height = height.lm$coefficients['x']*n.genes.plot+height.lm$coefficients[1])
                if (pdim[1] < min.width){pdim[1] <- min.width}

                fig(pdim['width.x'], pdim['height.x'])
                pdf(file = glue::glue('{save.dir}{name}.clustered.pdf'), width=pdim['width.x'], height=pdim['height.x'])
                plot(p)
                dev.off()
                ps[['clustered']] <- p
                
                p$data%>% select(feature.groups,features.plot) %>% arrange(feature.groups,desc(features.plot)) %>% unique() -> sorted_genes_last 
                write.csv(sorted_genes_last, glue::glue('{save.dir}{name}.clustered.genelist.csv'),row.names=F)

            }
        }
    }

    
    # without clustering
    if (without_clustering){
        message('--> Without clustering...')
        p <- dotplot(seu = seu, 
                     gene.list = gene.list, 
                     y.lab = name,
                     cluster.genes = F,
                     pct.min = pct.min,
                     n.samples.over.pct = n.samples.over.pct, 
                     repeated.do = repeated.do,
                     ...)
        
        # override order of genes
        glevels <- genelevels[[l]]
        glevels <- glevels[glevels%in%p$data$features.plot]
        p$data %>% mutate(features.plot = factor(as.character(features.plot), glevels)) -> p$data
        
        if (!is.null(p)){
            n.groups <- p$data[,names(p$facet$params$rows), drop=F]%>%unique()
            n.groups <- if(!any(is.null(dim(n.groups)))){nrow(n.groups)}else{length(n.groups)}
            n.genes.plot <- nrow(unique(p$data[,c('features.plot', 'feature.groups')])) + n.groups*0.7

            n_cts <- length(unique(seu$anno_cluster_fct))
            
            if(n.genes.plot>=1){

                row.chars <- p$data[,names(p$facet$params$rows), drop=F]%>%
                    apply(1, function(x){sum(nchar(as.character(x)))}) %>% max
                if (row.chars==1){row.chars <- 10}

                pdim <- c(
                    # width = width.lm$coefficients['x']*row.chars+width.lm$coefficients[1], 
                    width = width.lm$coefficients['x']*row.chars+width.lm$coefficients[1]*n_cts/6+.5, 
                    height = height.lm$coefficients['x']*n.genes.plot+height.lm$coefficients[1])
                if (pdim[1] < min.width){pdim[1] <- min.width}
                fig(pdim['width.x'], pdim['height.x'])

                pdf(file = glue::glue('{save.dir}{name}.pdf'), width=pdim['width.x'], height=pdim['height.x'])    
                plot(p)
                dev.off()
                ps[['not clust']] <- p
                
                p$data%>% select(feature.groups,features.plot) %>% arrange(feature.groups,desc(features.plot)) %>% unique() %>%
                    ungroup() %>% mutate(gene.order=as.numeric(forcats::fct_rev(features.plot)), abs.order=seq_along(along.with = features.plot)) -> sorted_genes_last
                if(any(!p$data$features.plot%in%unique(unlist(gene.list), recursive=T))){stop()}
                all.genes <- unique(unlist(gene.list, recursive=T))
                all.genes.df <- lapply(names(gene.list), function(x){data.frame(features.plot=gene.list[[x]], feature.groups=x)})%>%do.call(what='rbind')
                all.genes.df
                if(any(!p$data$features.plot%in%unique(all.genes.df$features.plot))){stop()}
                sorted_genes_last <- merge(all.genes.df, sorted_genes_last, by=c('features.plot','feature.groups'), all=T)%>%mutate(expressed=!is.na(abs.order)) %>% arrange(abs.order)
                write.csv(sorted_genes_last, glue::glue('{save.dir}{name}.genelist.csv'),row.names=F)
                
            }
        }
    }

    return(ps)
}
```
