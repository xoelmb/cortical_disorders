## Functions

```r
my.FindMarkers <- function(obj,
                           by.group='cluster',
                           ident.var='condition', ident.1=NULL, ident.2=NULL, same.ncells=T,
                           no.thresholds=T, all.genes=T, 
                           test.use='LR', mem.safe=F,
                           verbose=T, no.warning=T, 
                           parallel=T, workers=NULL, ...){
    
    # Check if test warnings should be displayed
    my.warnings <- if (no.warning){suppressWarnings} else {function(x){x}}
    
    if (parallel){
        library(future)
        if (!mem.safe){
            options(future.globals.maxSize=Inf)
        }
        # use multicore for parallel
        workers <- if (is.null(workers)){parallel::detectCores()-1} else {workers}
        plan("multicore", workers = workers)
    }
    
    # Create a dull variable for grouping if there is none
    if (is.null(by.group)){
        obj$by.group <- 'no group'
        by.group <- 'by.group'
    }
    
    # Get current ident.var if none is provided
    
    
    # Set group as identity for split
    Idents(obj) <- by.group
    # Split
    obj.list <- SplitObject(obj, split.by = "ident")
    # Assign condition as identity
    for (i in 1:length(obj.list)) {
        Idents(obj.list[[i]]) <- ident.var 
    }
    

    
    # Check extra pars
    extra.params <- list(...)
    if (no.thresholds){
        
        # Define parameters for no thresholds
        no.th.pars <- list(
            min.pct = -Inf, 
            logfc.threshold = -Inf,  
            min.cells.feature = 1,  
            min.cells.group = 1)                         

        for (p in names(no.th.pars)){
            if (!p%in%names(extra.params)){
                extra.params[[p]] <- no.th.pars[[p]]
            }}}
    
    # Decide features to use
    if (all.genes){
        if (!'features' %in% extra.params){
            extra.params[['features']] <- rownames(obj)
        }
    }
    
    # Iterate objects/groups
    res.list <- lapply(
        obj.list, function(my.obj){
            
            # get group
            g <- unique(my.obj@meta.data[,by.group])
            if (verbose){message(g)}
            # Get number of cells per group compared
            n_group <- table(my.obj@meta.data[, ident.var])
            max_cells <- if ( same.ncells ){ min(n_group) } else { Inf }
            
            # Try to find marker genes
            tryCatch(
                expr = {
                    x <- my.warnings(rlang::exec(
                        
                        Seurat::FindMarkers, 
                        
                        object = my.obj,                        
                        max.cells.per.ident = max_cells,

                        test.use = test.use, 
                        ident.1 = ident.1,
                        ident.2= ident.2,
                        
                        verbose=verbose,
                        
                        !!!extra.params))
                    
                    # Annotate some info on the results
                    x$ident.1 <- ident.1
                    x$ident.2 <- ident.2
                    x$ncells.1 <- n_group[ident.1]
                    x$ncells.2 <- n_group[ident.2]
                    # x$n.patient <- n_pac_str
                    x$test <- test.use
                    x$max.cells.per.ident <- max_cells
                    x$gene <- rownames(x)
                    
                    if (by.group!='by.group'){
                        x[,by.group] <- g
                    }
                    return(x %>% arrange(desc(avg_log2FC)))
                }, cond = function(cond){
                    print('Could NOT compute.')
                    print(as.character(cond))
                    return(NULL)
                }, error = function(cond){
                    print('Could NOT compute.')
                    print(as.character(cond))
                    return(NULL)
                })})
    
     # Remove empty results
    res.list <- res.list[!is.null(res.list)]
    
    # Bind results together
    res <- do.call(res.list, what='rbind')
    
    return(res)

}
```
